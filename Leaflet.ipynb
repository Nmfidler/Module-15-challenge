```python
import folium

# Create a map centered at [0, 0]
m = folium.Map(location=[0, 0], zoom_start=2)

# Add markers for each earthquake
earthquake_data = {
    "type": "FeatureCollection",
    "features": [
        {
            "type": "Feature",
            "properties": {
                "mag": 4.7,
                "place": "Fiji region",
                "time": 1715621044161,
                "url": "https://earthquake.usgs.gov/earthquakes/eventpage/us6000mybf",
                "magType": "mb",
                "title": "M 4.7 - Fiji region"
            },
            "geometry": {
                "type": "Point",
                "coordinates": [-179.5013, -21.2337]
            },
            "id": "us6000mybf"
        },
        {
            "type": "Feature",
            "properties": {
                "mag": 4.9,
                "place": "Kermadec Islands region",
                "time": 1715619952371,
                "url": "https://earthquake.usgs.gov/earthquakes/eventpage/us6000mybg",
                "magType": "mb",
                "title": "M 4.9 - Kermadec Islands region"
            },
            "geometry": {
                "type": "Point",
                "coordinates": [-176.6489, -27.8705]
            },
            "id": "us6000mybg"
        }
    ]
}

for feature in earthquake_data['features']:
    mag = feature['properties']['mag']
    place = feature['properties']['place']
    url = feature['properties']['url']
    folium.Marker(location=[feature['geometry']['coordinates'][1], feature['geometry']['coordinates'][0]],
                  popup=f"<b>Magnitude:</b> {mag}<br><b>Location:</b> {place}<br><a href='{url}' target='_blank'>USGS Event Page</a>").add_to(m)

# Add legend
legend_html = '''
     <div style="position: fixed; 
                 bottom: 50px; left: 50px; width: 150px; height: 90px; 
                 border:2px solid grey; z-index:9999; font-size:14px;
                 background-color: white;
                 ">
     <b>Magnitude Legend</b><br>
     <i style="background: #FF0000"></i> 4.5+<br>
     <i style="background: #FFA500"></i> 4.0-4.4<br>
     </div>
     '''

m.get_root().html.add_child(folium.Element(legend_html))

# Display the map with layer control
folium.map.LayerControl().add_to(m)

# Display the map
m
```
